# Read .csv file and change its form as (time, price, volume)

```python
import pandas as pd
import numpy as np
import datetime
import pytz
```

```python
# Standardize file to form (time, price, volume) for tick data, craete a new dataframe
def standard_tick_csv(file):
    # Construct new index datetime with (YYMMDD HH:MM:SS)
    file['DATE'] = file['DATE'].astype(str)
    file['TIME'] = file['TIME'].astype(str)
    time = file['DATE'] + ' ' + file['TIME']
    ttime = pd.to_datetime(time).astype(str)
    
    # Reserve price and size as price and volume 
    long = len(file)
    nprice = file['PRICE']
    nvolume = file['SIZE']
    
    #new_df2 index is date-time, cols = [price volume]
    new_df2 = pd.DataFrame(np.random.randn(long,2),columns=['price','volume'])
    new_df2.price = pd.Series(nprice)
    new_df2.volume = pd.Series(nvolume)
    new_df2.index = ttime
    new_df2.index.name = 'time'
    
    # save the dataframe to .csv
    # import os
    # new_df2.to_csv('C:/Users/hsy/Desktop/GlobalAI/VPIN/MIR/goog/'+'GOOG_tick.csv')
    
    return new_df2
```
